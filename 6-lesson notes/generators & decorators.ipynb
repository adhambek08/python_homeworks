## Creating Generators

```python
def simple_generator():
    print('Executed')
    yield 1
    print('Second yield')
    yield 2
    yield 3
    yield 4

val = simple_generator()
next(val)
```

```python
next(val)
```

```python
val = simple_generator()

for i in simple_generator():
    print(i)
```

```python
val = simple_generator()

while True:
    print(next(val))
```

```python
# example 2

def infinite_counter(start=0):
    while True:
        yield start
        start +=1

for i in infinite_counter():
    print(i)
    if i ==5:
        break
```

```python
# example 3
def fib(limit):
    a, b = 0, 1
    while b < limit:
        yield b
        a,b = b, a+b

fib_numbers = fib(100)
print(fib_numbers)

for i in fib_numbers:
    print(i)
```

```python
def rrrange(start, stop = None, step = 1):
    if stop is None:
        stop=start
        start = 0

    num = start
    while num < stop:
        yield num
        num += step

nums = rrrange(10, 20, 3)
print(nums)
for i in nums:
    print(i)
```

## Generator expressions

```python
# example 1
gen_expr = (x ** 2 for x in range(5))

for x in gen_expr:
    print(x)
```

```python
# example 2
sum_of_squares = sum(x ** 2 for x in range(10))
print(sum_of_squares)
```

## Decorators

- creating decorators
- applying decorators

```python
def print_hello():
    print('hello')

a = print_hello
a()
```

```python
def decorator1(func):
    print("Before")
    func()
    print("After")

decorator1(print_hello)
```

```python
# example 1
def decorator(func):
    def wrapper():
        print('Before function call')
        func()
        print('After function call')
    return wrapper
```

```python
var = decorator(print_hello)
type(var)
```

```python
var()
var.__name__
```

```python
@decorator
def say_hello():
    print("Hello!")

# teng

def say_hello():
    print("Hello!")

say_hello = decorator(say_hello)

say_hello()
```

```python
# example 2
def bold(func):
    def wrapper():
        return f"<b>{func()}</b>"
    return wrapper

@bold
def greet():
    return "Hello"

greet()
```

```python
# example 3
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(4)
def greet(name):
    print(f"Hello, {name}!")

greet('name')
```
