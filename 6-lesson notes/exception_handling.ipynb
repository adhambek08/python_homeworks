## Handling exceptions

```python
try:
    a = 1/0
    print(a)
except ZeroDivisionError:
    print("Bo'lib bo'lmaydi")
```

```python
try: 
    a = [1]
    a[1]
except IndexError:
    print("Index Error bo'ldi")
```

```python
try:
    a = [1]
    a[1]
except Exception: 
    print("Error occured")

# Faqat except deb yozish = BaseException
```

## Raising exceptions

```python
try: 
    raise ImportError
except ImportError:
    print("Caught an ImportError")
```

```python
raise Exception('something bad happened')
```

## Examining Exceptions

```python
try:
    import something
except ImportError as e:
    print(e.args)
    print(e)
    print(type(e))
```

## Using the finally statement

```python
try: 
    a = 1/0
except ZeroDivisionError:
    print("cannot divide")
else:
    print(a)
finally:
    print('finally executed')
```

```python
a = int(input())

try:
    b = 1/a
except ZeroDivisionError:
    pass
else:
    print(b+1)
```
