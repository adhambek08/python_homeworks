```python
def check(func):
    def wrapper(a, b):
        if b == 0:
            return "Denominator can't be zero"
        return func(a, b)
    return wrapper

@check
def div(a, b):
    return a / b

print(div(6, 2))  # 3.0
print(div(6, 0))  # Denominator can't be zero
```

```python
def add_employee():
    emp_id = input("Employee ID: ")
    name = input("Name: ")
    position = input("Position: ")
    salary = input("Salary: ")
    with open("employees.txt", 'a') as f:
        f.write(f'{emp_id}, {name}, {position}, {salary}\n ')
    print("✅ Employee added successfully!")

def view_all():
    try:
        with open("employees.txt", 'r') as f:
            records = f.readlines()
        if not records:
            print("No records found.")
        else:
            print("\n --- All employees ---")
            for record in records:
                print(record.strip())
    except FileNotFoundError:
        print("No employee file found yet.")

def search_employee():
    emp_id = input("Enter Employee ID to search: ").strip()
    try:
        with open('employees.txt', "r") as f:
            for line in f:
                if line.strip().startswith(emp_id + ","):
                    print("Found:", line.strip())
                    return
        print("Employee not found.")
    except FileNotFoundError:
        print("No employee file found yet.")

def update_employee():
    emp_id = input("Enter Employee ID to update: ")
    try:
        with open('employees.txt', "r") as f:
            records = f.readlines()

        updated = False
        new_records = []
        for line in records:
            if line.startswith(emp_id + ","):
                parts = line.strip().split(", ")
                print(f"Current: {line.strip()}")
                name = input(f"New Name ({parts[1]}): ") or parts[1]
                position = input(f"New Position ({parts[2]}): ") or parts[2]
                salary = input(f"New Salary ({parts[3]}): ") or parts[3]
                new_records.append(f"{emp_id}, {name}, {position}, {salary}\n")
                updated = True
            else:
                new_records.append(line)

        if updated:
            with open("employees.txt", "w") as f:
                f.writelines(new_records)
            print("✅ Employee updated successfully!")
        else:
            print("Employee not found.")
    except FileNotFoundError:
        print("No employee file found yet.")

def delete_employee():
    emp_id = input("Enter Employee ID to delete: ")
    try:
        with open("employees.txt", "r") as f:
            records = f.readlines()

        new_records = [line for line in records if not line.startswith(emp_id + ",")]

        if len(new_records) == len(records):
            print("Employee not found.")
        else:
            with open("employees.txt", "w") as f:
                f.writelines(new_records)
            print("✅ Employee deleted successfully!")
    except FileNotFoundError:
        print("No employee file found yet.")

def main():
    while True:
        print("\n--- Employee Records Manager ---")
        print("1. Add new employee record")
        print("2. View all employee records")
        print("3. Search for an employee by Employee ID")
        print("4. Update an employee's information")
        print("5. Delete an employee record")
        print("6. Exit")

        choice = input("Choose an option (1-6): ")

        if choice == "1":
            add_employee()
        elif choice == "2":
            view_all()
        elif choice == "3":
            search_employee()
        elif choice == "4":
            update_employee()
        elif choice == "5":
            delete_employee()
        elif choice == "6":
            print("Goodbye!")
            break
        else:
            print("Invalid option. Please choose 1-6.")

main()
```

```python
import string
from collections import Counter

SAMPLE_FILE = "namuna.txt"
REPORT_FILE = "word_count_report.txt"

# 1. File Input
try:
    with open(SAMPLE_FILE, "r") as f:
        text = f.read()
    print(f"Reading from '{SAMPLE_FILE}'...")
except FileNotFoundError:
    print(f"'{SAMPLE_FILE}' not found. Please type a paragraph:")
    text = input("> ")
    with open(SAMPLE_FILE, "w") as f:
        f.write(text)
    print(f"'{SAMPLE_FILE}' created!")

# 2. Count Word Frequency
# lowercase + punctuation olib tashlash
text = text.lower()
text = text.translate(str.maketrans("", "", string.punctuation))

words = text.split()
total_words = len(words)

word_counts = Counter(words)
top_5 = word_counts.most_common(5)

# 3. Output
print(f"\nTotal words: {total_words}")
print("Top 5 most common words:")
for word, count in top_5:
    time_label = "time" if count == 1 else "times"
    print(f"  {word} - {count} {time_label}")

# 4. Save to report file
with open(REPORT_FILE, "w") as f:
    f.write("Word Count Report\n")
    f.write(f"Total Words: {total_words}\n")
    f.write("Top 5 Words:\n")
    for word, count in top_5:
        f.write(f"{word} - {count}\n")

print(f"\n✅ Report saved to '{REPORT_FILE}'")
```
