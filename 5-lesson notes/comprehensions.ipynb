## List comprehensions ##

```python
numbers = [2,3,4,5,6]
a_square = []

for number in numbers:
    a_square.append(number **2)

print(a_square)
```

```python
numbers = [2,3,4,5,6]
a_square = []

for number in numbers:
    if number % 2 == 0:
        a_square.append(number **2)

print(a_square)
```

```python
#example1

numbers = [2,3,4,5,6]
a_squares = [num**2 for num in numbers]
print(a_squares)
```

```python
#example2
numbers = [2,3,4,5,6]
a_squares = [(num, num**2) for num in numbers if num % 2 == 0]
print(a_squares)
```

```python
a = []
for i in range(3):
    for j in range(3):
        a.append((i,j))
print(a)        
```

```python
# example 3

coordinates = [(x,y) for x in range(3) for y in range(3)]
coordinates
```

```python
# example 4
matrix = [[1,2,3], [4,5,6,], [7,8,9]]
flattend = [val for row in matrix for val in row]
print(flattend)
```

```python
matrix = ([1,2,3], [4,5,6,], [7,8,9])
flattened = []
for row in matrix:
    for item in  row:
        flattened.append(item)

print(flattend)
```

## Dictionary Comprehensions

```python
keys = ['a','b','c','d','e']
values = [1,2,3,4,5]

d = {}
for key, val in zip(keys, values):
    d[key]=val

# for i in range(len(keys)):
#     d[keys[i]] = values[i]

d 
```

```python
#example 1
keys = ['a','b','c','d','e']
values = [1,2,3,4,5]

d = {k:v for k,v in zip(keys, values)} # k:v => 2 nuqta yozilishi kerak
d
```

```python
# example 2
d_reverse = {v:k for k,v in d.items()}
d_reverse
```

```python
# example 3
numbers = [1,2,3,4,5]
squares = {num:num**2 for num in numbers}
squares
```

```python
# examaple 4
matrix = [[1,2,3], 
          [4,5,6], 
          [7,8,9]
]

matrix_squares = {pos: {num:num**2 for num in row} for pos, row in enumerate(matrix)}
matrix_squares
```

## Set Comprehensions


```python
letter = """Python is an interpreted, object-oriented,
high-level programming language with dynamic semantics.
Its high-level built in data structures, combined with
dynamic typing and dynamic binding, make it very attractive
for Rapid Application Development, as well as for use as
a scripting or glue language to connect existing components
together. Python's simple, easy to learn syntax emphasizes
readability and therefore reduces the cost of program
maintenance. Python supports modules and packages, which
encourages program modularity and code reuse."""

cleaned = letter.lower().replace('.', '')\
.replace(',','').replace('\n', '')\
.split()

print(cleaned)
```

```python
unique_words = []
for word in cleaned:
    if len(word)<3 and word not in unique_words:
        unique_words.append(word)

print(unique_words)
```

```python
unique_words = {word for word in cleaned if len(word)<3}
unique_words
```

## Generator exprressions
