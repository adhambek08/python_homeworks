## Arguments

## positional and keyword arguments

```python
# example 1
def greet(name, message):
    print(f'{message}, {name}!')

greet('Hello', 'Alice') #positional argument
```

```python
# example 2
greet(name = 'Bob', message = 'Hi')
greet(message = 'Welcome', name = 'Charlie')
```

```python
greet('Bob', message='Bob') # birinchi position keyin keyword keladi
```

## Positional-only va keyword-only arguments

```python
# example 1
def pos_only(x, /, y):
    return x+y

print(pos_only(3, y=4)) # 'x' must be positional
```

```python
# example 2: Keyword-only arguments using *
def key_only(*, x, y):
    return x/y

print(key_only(x=3, y=5)) # x and y must be keyword arguments
```

## Default arguments

```python
#example1

def greet(name='Guest'):
    print(f'Hello, {name}!')

greet()
greet('Alice')
```

```python
# example 2
def power(base, exponent = 2):
    return base ** exponent

print(power(3))
print(power(4,5))
```

## Arbitrary arguments

```python
def demo(a, *args, b):
    print(f'{a=}')
    print(f'{b=}')
    print(type(args))
    print(args)

demo(1,2,3,4,b=7)
```

```python
# example1 : Arbitrary positional arguments (*args)
def sum_all(*args):
    return sum(args)

print(sum_all(1,2,3,4,5))
```

```python
# example 2: arbitrary keyword arguments (**kwargs)
def print_info(**kwargs):
    print(type(kwargs))
    print(kwargs)

print_info(name = 'Alice', age=25, city = 'New York')
```

```python
def demo2(a,b, *args, **kwargs):
    print(args)
    print(kwargs)
    print(a,b)
    

demo2(1,2,45,5,7,3,2,22, f=1, d=6)
```

## Lambda expressions

```python
# example 2:
square = lambda x: x**2
square
```

```python
square(2)
```

```python
# example 2: Lambda with multiple arguments
add = lambda a,b: a+b
print(add(3,8))
```

```python
numbers = [1,2,3,4,5]

def square(x):
    return x ** 2

list(map(square, numbers))
```

```python
# *example 3: lambda used in high-order function (like map)*

numbers = [1,2,3,4,5]
squares = list(map(lambda x: x**2, numbers))
print(squares)
```

```python
# example 4: Sorting with Lambda
names = ['Charlie', 'Alice', 'Bob', 'David']
names.sort(key=lambda name: len(name), reverse = True)
print(names)
```

```python
# example 4: Sorting with Function

names = ['Charlie', 'Alice', 'Bob', 'David']

def get_length(name):
    return len(name)

names.sort(key=get_length, reverse=True)

print(names)
```

```python
names = ['Charlie', 'Alice', 'Bob', 'David']

def sort_by_length(name):
    return len(name)

names.sort(key=sort_by_length, reverse = True)
print(names)
```
