```python
list1 = [1, 1, 2]
list2 = [2, 3, 4]

common = set(list1) & set(list2)
result = []

for i in list1:
    if i not in common:
        result.append(i)

for i in list2:
    if i not in common:
        result.append(i)

print(result)
```

```python
txt = 'hello'

result = ""

for i in range(len(txt)):
    result += txt[i]

    if (i + 1) % 3 == 0:
        result += "_"

print(result)
```

```python
#Har 3 ta harfdan keyin _ qo'y, lekin 3-harf unli bo'lsa, _ ni keyingi harfdan keyin qo'y.

txt = 'assalom'

vowels = "aeiouAEIOU"
result = ""
count = 0
delay = False

for i in txt:
    result += i
    count += 1

    if delay:
        result += "_"
        delay = False
        count = 0

    elif count == 3:
        if i in vowels:
            delay = True
        else:
            result += "_"
            count = 0

if result.endswith("_"):
    result = result[:-1]

print(result)
```

```python
import random

while True:
    secret = random.randint(1, 100)
    attempts = 0
    guessed = False

    while attempts < 10:
        guess = int(input("Enter your guess (1-100): "))
        attempts += 1

        if guess > secret:
            print("Too high!")
        elif guess < secret:
            print("Too low!")
        else:
            print("You guessed it right!")
            guessed = True
            break

    if not guessed:
        answer = input("You lost. Want to play again? ")

        if answer in ["Y", "YES", "y", "yes", "ok"]:
            continue
        else:
            print("Thanks for playing!")
            break
    else:
        answer = input("Play again? ")

        if answer in ["Y", "YES", "y", "yes", "ok"]:
            continue
        else:
            print("Thanks for playing!")
            break
```

```python
password = input("Enter a password: ")

if len(password) < 8:
    print("Password is too short.")
elif not any(c.isupper() for c in password):
    print("Password must contain an uppercase letter.")
else:
    print("Password is strong.")
```

```python
for num in range(2, 101):
    is_prime = True

    for divisor in range(2, num):
        if num % divisor == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```

```python
txt = 'abcabcdabcdeabcdefabcdefg'

unlilar =  'aeiou'
counter = 0
javob = ""

for i in range(len(txt)):
    counter += 1
    javob += txt[i]
    if i!=len(txt)-1 and counter >= 3 and txt[i] not in unlilar:
        unlilar += txt[i]
        javob += '_'
        counter = 0

print(javob)
```

```python
txt = 'abcabcdabcdeabcdefabcdefg'

unlilar =  'aeiou'
counter = 0
javob = []

for i in range(len(txt)):
    counter += 1
    javob.append(txt[i])
    if i!=len(txt)-1 and counter >= 3 and txt[i] not in unlilar:
        unlilar += txt[i]
        javob.append('_')
        counter = 0

print(''.join(javob))
```
