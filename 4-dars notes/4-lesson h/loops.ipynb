```python
l = [1,2,3,]
i = 0
check_number = 3
i_list = []
while i < len(l):
  i_list.append(i)
  if l[i] == check_number:
    print('Found')
    break
  i += 1
else:
    print('Number not found')
print(i_list)
```

## For Loop - sequencelarni ustida amalga oshirish mumkin

```python
my_list = [1,2,3]
other_list = []
i = 0
while i < len(my_list):
    a = my_list[i]
    other_list. append(a**2)
    i +=1

print(other_list)    
```

```python
other_list = []
for a in my_list:
    other_list.append(a**2)

print(other_list)
```

```python
my_str = 'abcdefghi'
for char in my_str:
    print(char)
```

```python
users = {'user1':'John', 'user2': 'Adam', 'user3':'Josh'}

for user in users:
    print(user, users[user])
    
```

```python
users = {'user1':'John', 'user2': 'Adam', 'user3':'Josh'}

for user in users.values():
    print(user)
```

```python
for key, value in users.items():
    print(key, value)
```

```python
users = [('user1','John'), ('user2', 'Adam'), ('user3','Josh')] # List of tuples
for user in users:
    print(user)
```

```python
users = [('user1','John'), ('user2', 'Adam'), ('user3','Josh')] # List of tuples
for key, user in users:
    print(user) #faqat tupleda mumkin unpack qilish
```

```python
language = ['java', 'Python', 'JS']
versions = [14, 3, 6]

for i in range(len(language)):
    print(language[i], versions[i])
```

```python
# zip - listlarni mos indexlarini birlashtirish uchun ishlatiladi
for version, language in zip(versions, language):
    print(language, version)
```

```python
# For each value of i, run a full loop of j values.

for i in range(10):
    for j in range(10):
        print(i, j)
```
